```python
#function to resize the image to 360*240

def resize(image):
    img=cv2.imread(image)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    # First, ask the player about their CPU
    shape = len(img.shape)
    print(shape)
    if img.shape[:2] == (240,360):
        img=cv2.rotate(img,cv2.cv2.ROTATE_90_CLOCKWISE)
        return img
    if img.shape[0] <= (360) and img.shape[1] <=240:
        return img
    if img.shape[0]>360:
        img=cv2.resize(img,(360,img.shape[1]))
    if img.shape[1]>240:
        img=cv2.resize(img,((img.shape[0]),240))
    return img

    
```

### CREATING MOSAIC USING SLICING

```python
import cv2 as cv
img=cv.imread('../DATA/chessboard_mat.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
i,j,k=img.shape
ih,jh=int(i/2),int(j/2)
cpy=img.copy()
```

```python
img[:ih,:jh,:]=cpy[ih:i,jh:j,:]
img[ih:i,jh:j,:]=cpy[:ih,:jh,:]
img[ih:i,:jh,:]=cpy[:ih,jh:j,:]
img[:ih,jh:j,:]=cpy[ih:i,:jh,:]
```

```python
plt.imshow(img)  
```
